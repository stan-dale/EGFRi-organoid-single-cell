#### 6. Running Decoupler

##### Imports

```python
import sys
from pathlib import Path
_p = Path(".").resolve()
while not (_p / "src" / "config.py").exists() and _p != _p.parent:
    _p = _p.parent
sys.path.insert(0, str(_p))

import scanpy as sc
from src.config import DATASETS, FIGURES_DIR
from src.decoupler import run_decoupler_safe
```

```python
adata = sc.read_h5ad(DATASETS["D2_Lapa"]["labelled"])
```

```python
adata, score = run_decoupler_safe(adata, celltype_key="manual_label")
```

#### Set a save directory for our figures

```python
sc.settings.figdir = str(FIGURES_DIR / "decoupler/d2-dz")
```

```python
tf = "IRF1"
sc.pl.umap(score, color=[tf, "manual_label"], cmap="RdBu_r", vcenter=0, title=[f"{tf} score", "manual_label"], save="_irf1.pdf")
```
