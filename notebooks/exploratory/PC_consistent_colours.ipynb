#### Imports

```python
import sys, os
import scanpy as sc

parent_dir = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(parent_dir)

project_root = os.path.abspath(
    os.path.join(os.getcwd(), "..", "..")
)
sys.path.append(project_root)

from src.utils import celltype_palette
from pathlib import Path
sys.path.insert(0, str(Path("../..").resolve()))
from src.config import ANALYSIS_DIR, FIGURES_DIR
```

```python
#### Set our figure directory

sc.settings.figdir = str(FIGURES_DIR / "manual-label-umaps/v2-simplified-labels")   # or an absolute path
```

#### Load data

```python
d2_dz = sc.read_h5ad(ANALYSIS_DIR / "d2_dz_manual_labels_2.h5ad")
```

#### Plot both representations

```python
adata = d2_dz
```

```python
key = "pc_manual_label"  # the obs column you're plotting

# make sure it's categorical
adata.obs[key] = adata.obs[key].astype("category")

# assign colours in the order of the categories
adata.uns[f"{key}_colors"] = [
    celltype_palette[ct] for ct in adata.obs[key].cat.categories
]
```

```python
sc.pl.umap(d2_dz, color = 'pc_manual_label', title="Day 2", save="d2_dz_v1.pdf")
```

```python
key = "manual_label"  # the obs column you're plotting

# make sure it's categorical
adata.obs[key] = adata.obs[key].astype("category")

# assign colours in the order of the categories
adata.uns[f"{key}_colors"] = [
    celltype_palette[ct] for ct in adata.obs[key].cat.categories
]
```
