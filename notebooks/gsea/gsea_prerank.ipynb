Set `COMPARISON` below to any key from `dge_comparisons.yaml`, then run all cells.

**Environment:** `pydeseq-env`

```python
# ── Configuration ──
# Change this to run a different comparison:
COMPARISON = "d2_lapa_iscs"
```

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path("../..").resolve()))

from src.config import DGE_OUTPUT_DIR, GSEA_OUTPUT_DIR
from src.gsea import GSEAEnrichmentAnalyzer
from src.markers import pathway_gene_sets
```

```python
# ── Derive paths from comparison key ──
dge_file = DGE_OUTPUT_DIR / f"{COMPARISON}_deseq2_results.csv"
output_dir = GSEA_OUTPUT_DIR / COMPARISON

print(f"DGE input:   {dge_file}")
print(f"GSEA output: {output_dir}")
assert dge_file.exists(), f"DGE results not found: {dge_file}"
```

```python
# ── Run full GSEA analysis ──
analyzer = GSEAEnrichmentAnalyzer(
    dge_file=dge_file,
    output_dir=output_dir,
    name=COMPARISON,
    gene_sets=pathway_gene_sets,
)
analyzer.run_full_analysis()
```

```python
# ── Results table ──
analyzer.gsea_results.res2d
```

```python
# ── Summary statistics ──
res = analyzer.gsea_results.res2d
n_sig = (res["FDR q-val"].astype(float) < 0.25).sum()
print(f"Pathways tested: {len(res)}")
print(f"Significant (FDR < 0.25): {n_sig}")
```
