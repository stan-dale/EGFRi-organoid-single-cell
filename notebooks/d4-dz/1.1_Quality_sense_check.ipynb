```python
## Imports
import scanpy as sc
import numpy as np
import scipy as sp
import os, sys


from pathlib import Path
sys.path.insert(0, str(Path("../..").resolve()))
from src.config import ANALYSIS_DIR, DATA_DIR, FIGURES_DIR
```

```python
## Load data
d4_dz = sc.read_h5ad(DATA_DIR / "egfDuod_D4_DZ.h5ad")
```

```python
## Identify our normalised and raw matrices
adata = d4_dz
sc.settings.figdir = str(FIGURES_DIR / "qc/d4-dz")
```

```python
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.pp.highly_variable_genes(adata, n_top_genes=2000)
adata = adata[:, adata.var.highly_variable]
sc.pp.scale(adata, max_value=10)
sc.tl.pca(adata, svd_solver='arpack')
sc.pp.neighbors(adata, n_neighbors=30, n_pcs=30)
sc.tl.leiden(adata, resolution=1, n_iterations = -1)
sc.tl.umap(adata)
```

```python


sc.pl.pca(adata, color="leiden")
sc.pl.umap(adata, color="leiden")

sc.pl.umap(adata, color="leiden", save="_d4_dz_clustered.pdf")
```

```python
#### This is much better
d4_dz.obsm['X_pca'] = adata.obsm['X_pca']
d4_dz.obsm['X_umap'] = adata.obsm['X_umap']
d4_dz

d4_dz.obs['leiden'] = adata.obs['leiden'].astype('category')
d4_dz.obsp['distances'] = adata.obsp['distances']
d4_dz.obsp['connectivities'] = adata.obsp['connectivities']
```

```python
intermediate_directory = str(ANALYSIS_DIR)
sc.write(f'{intermediate_directory}/re_clustered_egfDuod_D4_DZ.h5ad', d4_dz)
```
